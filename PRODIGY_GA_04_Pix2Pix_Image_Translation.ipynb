```python
import torch
import diffusers

print("Libraries Installed Successfully")
print("Diffusers Version:", diffusers.__version__)
```

```python
from diffusers import StableDiffusionInstructPix2PixPipeline
import torch

pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
    "timbrooks/instruct-pix2pix",
    torch_dtype=torch.float16,
    safety_checker=None
)

pipe = pipe.to("cuda")

print("Pix2Pix Model Loaded Successfully")
```

```python
from PIL import Image

image = Image.open("task_4.jpg").convert("RGB")

edited_image = pipe(
    prompt="Turn this image into a beautiful cartoon style artwork",
    image=image,
    num_inference_steps=20,
    image_guidance_scale=1.5
).images[0]

edited_image.save("translated_image.png")

edited_image
```

```python
from PIL import Image
import matplotlib.pyplot as plt

original = Image.open("task_4.jpg")
generated = Image.open("translated_image.png")

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(original)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(generated)
plt.title("Pix2Pix Output")
plt.axis("off")

plt.show()
```
